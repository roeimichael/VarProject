```python
net_liquidity = 294000
df = pd.read_csv('./data/actualportfolio.csv')
df
```

```python
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
df = df.drop(['Daily P&L','Last','Change','Unrealized P&L', 'Market Value'],axis = 1)
df.rename(columns = {'Financial Instrument':'Symbol'}, inplace = True)
df['Sector'] = ''
df['Amount'] = 0.0
for index, row in df.iterrows(): 
    df.at[index,'Position'] = df.at[index,'Position'].replace(',', '')
    ticker = df.at[index,'Symbol']
    tick = yf.Ticker(ticker)
    if 'sector' in tick.info:
        df.at[index,'Sector'] = tick.info['sector']
    else :
         df.at[index,'Sector'] = 'ETF'
df["Position"] = pd.to_numeric(df["Position"])          
df['Amount'] = df['Position']* df['Avg Price']  
df['Type'] = np.where(df['Amount'] >0 , 'LONG', 'SHORT')
df = df.drop(['Avg Price','Position'], axis = 1)
df['Percentage'] = abs(df['Amount'])/net_liquidity
cols = df.columns.tolist() 
cols = ['Symbol','Type','Amount' , 'Percentage', 'Sector']
df = df[cols]
df.rename(columns = {'Type':'Position', 'Percentage':'Protfilio Precentage'}, inplace = True)
df
```

```python

def get_ticker_pairs():
    tickers = []
    with open('ticker-pairs.txt', 'r') as fp:
        for line in fp:
            x = line[:-1]
            tickers.append(x)
    return tickers

def get_tickers():
    tickers = []
    with open('tickers.txt', 'r') as fp:
        for line in fp:
            x = line[:-1]
            tickers.append(x)
    return tickers
```

```python
tickers = get_tickers()
for ticker in tickers:
    print(ticker)
    stock = yf.Ticker(ticker)
    df = stock.history(start='2021-04-26', end='2023-02-26', interval="1h", prepost=False)
    df['ma50'] = df['Open'].rolling(50).mean()
    df['RSI'] = ta.RSI(df['Close'], timeperiod=14)
    df['ATR'] = ta.ATR(df['High'], df['Low'], df['Close'], timeperiod=14)
    df['BBupperband'], df['BBmiddleband'], df['BBlowerband'] = ta.BBANDS(df['Close'], timeperiod=5, nbdevup=2, nbdevdn=2, matype=0)
    df = df.drop(columns=['Stock Splits', 'Dividends'])
    df.to_csv(f"./data/stocks/plain/{ticker}.csv")
```

```python
ticker_pairs = get_ticker_pairs()
for pair in ticker_pairs:
    ticker_x = pair.split("/")[0]
    ticker_y = pair.split("/")[1]
    stock_x = yf.Ticker(ticker_x)
    stock_y = yf.Ticker(ticker_y)
    df_x = stock_x.history(start='2021-04-26', end='2023-02-26', interval="1h", prepost=False)
    df_y = stock_y.history(start='2021-04-26', end='2023-02-26', interval="1h", prepost=False)
    df = df_x/df_y
    df['ma50'] = df['Open'].rolling(50).mean()
    df['RSI'] = ta.RSI(df['Close'], timeperiod=14)
    df['ATR'] = ta.ATR(df['High'], df['Low'], df['Close'], timeperiod=14)
    df['BBupperband'], df['BBmiddleband'], df['BBlowerband'] = ta.BBANDS(df['Close'], timeperiod=5, nbdevup=2, nbdevdn=2, matype=0)
    df = df.drop(columns=['Stock Splits', 'Dividends'])
    df.to_csv(f"./data/stocks/pairs/{ticker_x}-{ticker_y}.csv")
```
